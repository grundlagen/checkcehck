```python
!pip install transformers torchaudio numpy
from transformers import Wav2Vec2Model
import torch
import torchaudio
import numpy as np
from google.colab import files

# Load the pre-trained Wav2Vec 2.0 model
model = Wav2Vec2Model.from_pretrained("facebook/wav2vec2-base-960h")

# Function to preprocess audio
def preprocess_audio(file_path):
    waveform, sample_rate = torchaudio.load(file_path)
    if sample_rate != 16000:
        resampler = torchaudio.transforms.Resample(orig_freq=sample_rate, new_freq=16000)
        waveform = resampler(waveform)
    return waveform.squeeze()  # Remove any extra dimensions

# Function to extract features
def extract_features(model, waveform):
    with torch.no_grad():
        outputs = model(waveform.unsqueeze(0))  # Add batch dimension
    return outputs.last_hidden_state.squeeze(0)  # Remove batch dimension

# List of your audio files
audio_files = ["common_voice_fr_38487438.mp3",
    "common_voice_fr_38487439.mp3",
    "common_voice_fr_38487440.mp3",
    "common_voice_fr_38487441.mp3",
    "common_voice_fr_38487442.mp3",
    "common_voice_fr_38487444.mp3",
    "common_voice_fr_38487445.mp3",
    "common_voice_fr_38487446.mp3",
    "common_voice_fr_38487451.mp3",
    "common_voice_fr_38487452.mp3"]


for file in audio_files:
    # Preprocess the audio
    waveform = preprocess_audio(file)

    # Check if waveform is empty
    if waveform.numel() == 0:
        print(f"Skipping {file} due to empty waveform.")
        continue

    # Extract features
    features = extract_features(model, waveform)

    # Save features to a file (numpy format)
    npy_file = file.replace('.mp3', '_features.npy')
    np.save(npy_file, features.numpy())

    print(f"Processed and saved features for {file}")

    # Initiate download of the numpy file
    files.download(npy_file)
```

```python
import tarfile
import os
import shutil

def extract_and_move(tar_file_path, extraction_directory, destination_directory):
    # Create the extraction directory if it doesn't exist
    os.makedirs(extraction_directory, exist_ok=True)

    # Try to open and extract the tar file
    try:
        with tarfile.open(tar_file_path, 'r') as tar:
            tar.extractall(path=extraction_directory)
            print("Extraction complete.")
    except Exception as e:
        print(f"Error during extraction: {e}")
        return

    # Function to recursively copy directories
    def copy_directory(src, dst):
        os.makedirs(dst, exist_ok=True)
        for item in os.listdir(src):
            s = os.path.join(src, item)
            d = os.path.join(dst, item)
            if os.path.isdir(s):
                copy_directory(s, d)
            else:
                shutil.copy2(s, d)

    # Manually move the files and directories to the desired Google Drive folder
    try:
        for item in os.listdir(extraction_directory):
            source_path = os.path.join(extraction_directory, item)
            destination_path = os.path.join(destination_directory, item)

            if os.path.isdir(source_path):
                # Copy the directory recursively
                copy_directory(source_path, destination_path)
            else:
                # Copy the file
                shutil.copy2(source_path, destination_path)

            print(f"Copied {item} to {destination_path}")

            # Remove the item (file or directory) from the source directory
            if os.path.isdir(source_path):
                shutil.rmtree(source_path)
            else:
                os.remove(source_path)

    except Exception as e:
        print(f"Error during file moving: {e}")
        return

    print("All files and directories have been copied and removed successfully.")

# Path to the tar file
tar_file_path = '/content/drive/MyDrive/comun voces/cv-corpus-16.1-delta-2023-12-06-en.tar'

# Directory where you want to extract the files
extraction_directory = '/content/extracted/'

# Destination directory on Google Drive
destination_directory = '/content/drive/MyDrive/comun voces/'

# Call the function with the specified parameters
extract_and_move(tar_file_path, extraction_directory, destination_directory)
```

```python
import tarfile
import os

def extract_limited_size(tar_file_path, destination_directory, max_size_mb=700):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_directory, exist_ok=True)

    # Convert max size in MB to bytes
    max_size_bytes = max_size_mb * 1024 * 1024

    # Try to open and partially extract the tar file
    try:
        with tarfile.open(tar_file_path, 'r') as tar:
            current_size = 0
            for member in tar.getmembers():
                # Check if adding this file will exceed the limit
                if current_size + member.size > max_size_bytes:
                    break
                tar.extract(member, path=destination_directory)
                current_size += member.size
            print(f"Partial extraction complete. Extracted up to {current_size / (1024 * 1024):.2f} MB")
    except Exception as e:
        print(f"Error during extraction: {e}")

# Path to the tar file
tar_file_path = '/content/drive/MyDrive/comun voces/cv-corpus-16.1-delta-2023-12-06'

# Destination directory on Google Drive where you want the files to be extracted
destination_directory = '/content/drive/MyDrive/comun voces/'

# Call the function with the specified parameters
extract_limited_size(tar_file_path, destination_directory)
```
