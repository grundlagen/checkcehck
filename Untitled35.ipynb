```python
!pip install -q -U numpy==1.24.4
!pip install -q -U TTS torch

# Step 2: Import libraries
import os
import numpy as np
import torch
from TTS.api import TTS
from TTS.tts.configs.xtts_config import XttsConfig
from TTS.config.shared_configs import BaseDatasetConfig
from google.colab import drive

# Step 3: Configure PyTorch
torch.serialization.add_safe_globals([XttsConfig, BaseDatasetConfig])
os.environ["TORCH_FORCE_NO_WEIGHTS_ONLY_LOAD"] = "1"

# Step 4: Mount Google Drive (if your files are stored there)
drive.mount('/content/drive')

# Step 5: Configure paths (UPDATE THESE TO YOUR ACTUAL PATHS)
# For files in Google Drive:
MODEL_DIR = "/content/drive/MyDrive/xtts_model_v2"  # Path to your model directory
WAV_DIR = "/content/common_voice_fr_39586341.wav"    # Path to your audio files
OUT_DIR = "/content/xtts embed"  # Output path

# For files uploaded directly to Colab:
# MODEL_DIR = "/content/xtts_model_v2"
# WAV_DIR = "/content/500wav_french"
# OUT_DIR = "/content/xtts_french_embed"

# Create output directory
os.makedirs(OUT_DIR, exist_ok=True)

# Step 6: Download model files if needed (uncomment if your model is online)
# !wget -P $MODEL_DIR https://example.com/model.pth
# !wget -P $MODEL_DIR https://example.com/config.json
# !wget -P $MODEL_DIR https://example.com/speakers_xtts.pth
# !wget -P $MODEL_DIR https://example.com/vocab.json

# Step 7: Load the model
tts = TTS()
tts.load_model(
    model_path=os.path.join(MODEL_DIR, "model.pth"),
    config_path=os.path.join(MODEL_DIR, "config.json"),
    speakers_file=os.path.join(MODEL_DIR, "speakers_xtts.pth"),
    vocab_path=os.path.join(MODEL_DIR, "vocab.json"),
    use_cuda=torch.cuda.is_available()  # Automatically use GPU if available
)

# Step 8: Extract embeddings
def extract_embeddings(wav_dir, out_dir):
    for file in os.listdir(wav_dir):
        if file.endswith(".wav"):
            filepath = os.path.join(wav_dir, file)
            try:
                # Compute conditioning latents
                conditioning_latents = tts.synthesizer.tts_model.get_conditioning_latents(
                    audio_path=filepath
                )

                # Extract speaker embedding
                speaker_embedding = conditioning_latents['speaker_embedding']

                # Convert to numpy and save
                embedding_np = speaker_embedding.detach().cpu().numpy()
                out_path = os.path.join(out_dir, os.path.splitext(file)[0] + ".npy")
                np.save(out_path, embedding_np)

                print(f"✅ Processed {file}")

            except Exception as e:
                print(f"❌ Error processing {file}: {str(e)}")

# Step 9: Run the extraction
# Assuming WAV_DIR points to a directory of WAV files.
# If WAV_DIR is a single file path, you might need to adjust this step.
# If WAV_DIR is a directory, uncomment the line below to run the extraction.
# extract_embeddings(WAV_DIR, OUT_DIR)

# Note: The original code had 'print' here without arguments.
# Assuming you intended to add something here, or it was a placeholder.
# If you want to run the extraction, uncomment the call to extract_embeddings above.
```
