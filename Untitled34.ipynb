```python
# XTTS Audio Embedding Matching Pipeline for Google Colab (Updated for Batch Matching)

# Step 1: Install Dependencies
!pip uninstall -y pandas numpy networkx
!pip install pandas==2.2.2 numpy==1.26.4 networkx==3.2 faiss-cpu --quiet
!pip install TTS==0.18.0 --quiet

# Step 2: Import Libraries
import os
import numpy as np
from TTS.api import TTS
import faiss
from tqdm import tqdm
import pandas as pd

# Step 3: Initialize XTTS Model
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")

# Step 4: Define Helper Function to Extract Embeddings from a Directory
def extract_embeddings_from_dir(wav_dir):
    embeddings = []
    files = []
    for f in tqdm(sorted(os.listdir(wav_dir))):
        if f.endswith(".wav"):
            wav_path = os.path.join(wav_dir, f)
            try:
                emb = tts.synthesizer.tts_model.get_conditioning_latent(wav_path)
                embeddings.append(emb)
                files.append(f)
            except Exception as e:
                print(f"Failed on {f}: {e}")
    return np.vstack(embeddings), files

# Step 5: Set Audio Folder Paths
eng_dir = "/bin/eng"
fr_dir = "/bin/fre"

# Step 6: Extract Embeddings from English and French Directories
eng_embs, eng_files = extract_embeddings_from_dir(eng_dir)
fr_embs, fr_files = extract_embeddings_from_dir(fr_dir)

# Step 7: Match Embeddings using FAISS
dim = eng_embs.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(fr_embs)
D, I = index.search(eng_embs, 1)

# Step 8: Save Matching Results to CSV
df = pd.DataFrame({
    "English_WAV": eng_files,
    "Best_Matching_French_WAV": [fr_files[i[0]] for i in I],
    "Distance": [d[0] for d in D]
})
df.to_csv("matched_wavs.csv", index=False)
print("\nMatching results saved as matched_wavs.csv")
```
